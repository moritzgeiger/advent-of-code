```python
import re
source = open('input2.txt').read()
mapper = {
    r"(1|one)": 1,
    r"(2|two)": 2,
    r"(3|three)": 3,
    r"(4|four)": 4,
    r"(5|five)": 5,
    r"(6|six)": 6,
    r"(7|seven)": 7,
    r"(8|eight)": 8,
    r"(9|nine)": 9
} 
```

```python
sum = 0
test_str = source

for i in test_str.split('\n'):
    print(f'before: {i=}')
    _helper = {}
    for k, v in mapper.items():
        try:
            for _i in re.finditer(k, i):
                _helper[_i.start()] = v
        except:
            continue
    print(f'after: {_helper=}')

    _srt = list(dict(sorted(_helper.items())).values())
    print(_srt)
    
    nums = int(f'{_srt[0]}{_srt[-1]}')
    print(f'{nums=}')
    sum += nums
```
